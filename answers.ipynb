# Problem 2

```python
vals = np.linspace(-3,3,200)

f = Sin()
f.plot(vals=vals, label='sin')
for deg in [0,1,3,5]:
    Tf = f.taylor_series(0, deg=deg)
    Tf.plot(vals=vals, label='degree {}'.format(deg))


plt.legend()
plt.show()
```
