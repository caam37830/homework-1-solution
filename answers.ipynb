## Part F: Synthesis Questions

1. What is the derivative of `5x^2 + 3x + 1`? 

```python
from functions import Polynomial
p = Polynomial(5, 3, 1)
p
```

```python
p('x')
```

```python
p.derivative()('x')
```

2. A rule for the derivative of $f = g/h$ using symbolic functions.  Does this reduce to quotient rule?

```python
from functions import Symbolic
g = Symbolic('g')
h = Symbolic('h')
f = g / h
print(f.derivative())
```

```python
print(f)
```

3. What is the derivative of $\sin(x)^2$?

```python
from functions import Sin, Polynomial
f = Polynomial(1,0,0)(Sin())
f('x')
```

```python
f.derivative()('x')
```

4. What is the second derivative of $e^{5x}$?

```python
from functions import Exponential, Scale
f = Exponential()(Scale(5))
f('x')
```

```python
f.derivative().derivative()('x')
```
